```python
# Задача 2. Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.
numbers_list = [] # объявляем пустой список
number_input = int(input('Введите число: '))
while number_input != 0: # обьявляем цикл вводить числа пока введенное число не будет равно 0
    numbers_list.append(number_input) # добавляем введенные числа в список
    number_input = int(input('Введите число: ')) 
x = sum(numbers_list) # когда введенное число равно 0 выходим из цикла и суммируем числа в списке
print(x)
```

```python
# Задача 3. Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек. Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки! Но мы не будем никого знакомить, если кто-то может остаться без пары.
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
boys_sorted = sorted(boys) # создаем новый список но уже с отсортированными именами
girls_sorted = sorted(girls)
if len(boys) != len(girls): # проверка на одинаковое количество имен
    print('Внимание, кто-то может остаться без пары!')
else:
    print('Идеальные пары:')
    x = 0
    while x < len(boys): # построчно печатаем имена из сортированных списков пока они не закончатся
        print(boys_sorted[x] + ' и ' + girls_sorted[x])
        x = x + 1
```

```python
# Звадача 4. У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере). Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
average_temperature = [] # обьявляем новый пустой лист куда будем записывать средние температуры
at = 0 # обьявляем переменную для использования в цикле
while at < len(countries_temperature): # запускаем цикл добавления средней температуры в лист со средними температурами
    average_temperature.append(round((sum(countries_temperature[at][1]) / len(countries_temperature[at][1]) - 32) * 5 / 9, 1)) # добавляем уже пересчитанные из фаренгейта в цельсии средние температуры в лист
    print(countries_temperature[at][0] + ' - ' + str(average_temperature[at]) + ' C') # построчно выводим значения из списка
    at = at + 1
```

```python
# Задача 5. Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
split_list = [] # Обьявляем пустой список в котором будем записывать разделенные запятыми значения
add_list = [] # Обьявляем список где будут храниться элементы разделенных по разделителю (,) значений первоначального списка
sp = 0 # обьявляем переменную для использования в цикле
while sp < len(stream):
    add = str(stream[sp]) # переводим значение очередного элемента списка в формат строки для дальнейшего разделения по разделителю
    add_list = add.split(',') # разделяем полученную строку по разделителю (,)
    split_list.append(add_list) # добавляем полученную строку в разделенном формате в список
    sp = sp + 1 
user_list = [] # Обьявляем список в котором будут храниться все имена пользователей
views_list = [] # Обьявляем список в котором будут храниться количества просмотров
ps = 0 # обьявляем переменную для использования в цикле
while ps < len(split_list):
    user_list.append(split_list[ps][1]) # переносим значения пользователей из разделенного списка в новый список
    views_list.append(int(split_list[ps][2])) # переносим значения количества просмотров из разделенного списка в новый список
    ps = ps + 1
views_sum = sum(views_list) # суммируем количества показов в списке
user_list_sorted = sorted(user_list) # сортируем список пользователей по порядку чтобы посчитать сколько повторяющихся пользователей а сколько уникальных
counter = 0
count = 0 # обьявляем переменную для использования в цикле
while count < len(user_list_sorted): # обявляем цикл в котором посчитаем количество уникальных пользователей, каждого нового уникального пользователя будем добавлять в счетчик
    if count == 0: # первый пользователь всегда уникальный
        counter = counter + 1 # добавляем в общий счетчик первого пользователя
        count = count + 1
    elif user_list_sorted[count] == user_list_sorted[count-1]: # если значение следующего пользователя совпадает с предыдущим то ничего не меняем и продолжаем дальше
        count = count + 1
    elif user_list_sorted[count] > user_list_sorted[count-1]: # если следующий пользователь отличается от предыдущего то прибавляем 1 в счетчик 
        counter = counter + 1
        count = count + 1
average = round((views_sum / counter), 2) # считаем среднее значение просмотров, округляем как в примере
print('Среднее количество просмотров на уникального пользователя: ' + str(average))
```
